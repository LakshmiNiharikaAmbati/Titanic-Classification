```python
test_data = pd.read_csv('/content/drive/MyDrive/test.csv')
train_data = pd.read_csv('/content/drive/MyDrive/train.csv')
```

preprocessing

```python
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
```

```python
train_data['Title'] = train_data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
# Encoding categorical variables
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked', 'Title'], drop_first=True)
```

```python
# Selecting relevant features and target
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S', 'Title_Miss', 'Title_Mr', 'Title_Mrs']  # Modify with your relevant features
target = 'Survived'

# Separating features and target variable
X_train = train_data[features]
y_train = train_data[target]
```

```python
# Handling missing values
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
test_data['Title'] = test_data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test_data = pd.get_dummies(test_data, columns=['Sex', 'Embarked', 'Title'], drop_first=True)
# Selecting relevant features
X_test = test_data[features]
```

Model Building

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Splitting the training data for model validation
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Training the model
model.fit(X_train_split, y_train_split)

# Predicting on the validation set
val_predictions = model.predict(X_val)

# Evaluating the model
print("Accuracy:", accuracy_score(y_val, val_predictions))
print("\nClassification Report:\n", classification_report(y_val, val_predictions))
print("\nConfusion Matrix:\n", confusion_matrix(y_val, val_predictions))
```

Model Evaluation

```python
# Predicting on the test set
test_predictions = model.predict(X_test)
result_df = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': test_predictions})
result_df.to_csv('predicted_survival.csv', index=False)
```
